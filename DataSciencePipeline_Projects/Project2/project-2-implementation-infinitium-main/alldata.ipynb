```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime
 

file="rdt_time.csv"


def remove_dot_zeros_from_df(file):
   df=pd.read_csv(file)
   df=df.fillna(0)
   dfcolumnlist=df.columns
   for column in dfcolumnlist:
       try:
           df[column]=df[column].astype(int)
       except Exception as e:
           print(e)
   return df    

data=remove_dot_zeros_from_df(file)


ans=data.sort_values(by=['_id.month','_id.day','_id.year','_id.hour']).reset_index(drop=True)
xvalues=[]
xvalues2=[]
yvalues=[]
for i, j in ans.iterrows():
    #print (str(j["_id.year"])+str(j["_id.month"]).zfill(2)+str(j["_id.day"]).zfill(2)+str(j["_id.hour"]).zfill(2))  
    #if(str(j["_id.year"])=="2022" and j["_id.day"] >=1 and j["_id.month"]==11 ):
    xvalues.append(str(j["_id.year"])+str(j["_id.month"]).zfill(2)+str(j["_id.day"]).zfill(2)+str(j["_id.hour"]).zfill(2))  
    yvalues.append (j["count"])
    datenew=(str(j["_id.year"])+str(j["_id.month"]).zfill(2)+str(j["_id.day"]).zfill(2)+str(j["_id.hour"]).zfill(2))  
    xvalues2.append(datetime.strptime(datenew, '%Y%m%d%H'))

plt.title('hourly collection of reddit comments',fontsize = 7)
plt.xlabel("days",fontsize = 7)
plt.ylabel("data collected ",fontsize = 7)
plt.tight_layout()
plt.bar(xvalues,yvalues)
plt.xticks(rotation=90,fontsize=3)
plt.yticks(rotation=90,fontsize=6)
plt.savefig("testreddit.png",bbox_inches='tight')
plt.tight_layout()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime
 

file="twt_daily.csv"


def remove_dot_zeros_from_df(file):
   df=pd.read_csv(file)
   df=df.fillna(0)
   dfcolumnlist=df.columns
   for column in dfcolumnlist:
       try:
           df[column]=df[column].astype(int)
       except Exception as e:
           print(e)
   return df    

data=remove_dot_zeros_from_df(file)


ans=data.sort_values(by=['_id.month','_id.day','_id.year']).reset_index(drop=True)
xvalues=[]
xvalues2=[]
yvalues=[]
for i, j in ans.iterrows(): 
    if(str(j["_id.year"])=="2022" and j["_id.day"] >=1 and j["_id.day"] <=14 and j["_id.month"]==11 ):
      xvalues.append(str(j["_id.year"])+str(j["_id.month"]).zfill(2)+str(j["_id.day"]).zfill(2))
      yvalues.append (j["count"])
      datenew=(str(j["_id.year"])+str(j["_id.month"]).zfill(2)+str(j["_id.day"]).zfill(2)) 
      xvalues2.append(datetime.strptime(datenew, '%Y%m%d'))

plt.title('daily collection of tweets Nov 1 - Nov 14',fontsize = 7)
plt.xlabel("days",fontsize = 7)
plt.ylabel("data collected ",fontsize = 7)
plt.tight_layout()
plt.bar(xvalues,yvalues)
plt.xticks(rotation=90,fontsize=6)
plt.yticks(rotation=90,fontsize=6)
plt.savefig("dailytwitter.png",bbox_inches='tight')
plt.tight_layout()
```
